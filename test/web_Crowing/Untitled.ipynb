```python
# step1.라이브러리 불러오기
import requests
from bs4 import BeautifulSoup as bs
import telegram
import time
import pandas as pd

# step2.제품명과 제품가격정보를 담을 빈 리스트 선언
type=[]
name=[]
eng_name=[]
degree=[]

# step3.for문을 이용해서 원하는 페이지에 접근, 정보 추출 후 리스트에 담기
for page_num in range(250):
    # range를 이용하면 0부터 인덱스가 시작되므로 page_num에 1을 더해준 url을 이용
    url = f'https://www.masileng.com/ingredient/{page_num}'
    
    # html 정보 받아와서 파싱
    response = requests.get(url)
    soup = bs(response.text , 'html.parser')

    # css selector로 페이지 내의 원하는 정보 가져오기
   
    html_type = soup.select('#__next > div.bubbling-limit > main > div > div.style_container__ELAz6 > div > div.style_description-container___qdtu > div.text__Text-sc-1y4jww2-0.bMFrhs.detail-item__HashTag-sc-c6hn19-3.FFTrJ')
    html_name = soup.select('#__next > div.bubbling-limit > main > div > div.style_container__ELAz6 > div > div.style_description-container___qdtu > div.style_title-wrap___gW30 > div.text__Text-sc-1y4jww2-0.bMFrhs.detail-item__Name-sc-c6hn19-0.gnNOfZ')
    html_eng_name = soup.select('#__next > div.bubbling-limit > main > div > div.style_container__ELAz6 > div > div.style_description-container___qdtu > div.style_title-wrap___gW30 > div.text__Text-sc-1y4jww2-0.bMFrhs.detail-item__EnglishName-sc-c6hn19-1.iItHEu')
    html_degree = soup.select('#__next > div.bubbling-limit > main > div > div.style_container__ELAz6 > div > div.style_description-container___qdtu > div.ingredient-detail__Flex-sc-jkj7o5-2.jlMHWx.sub-wrap > div')
    
    # 텍스트만 추출
    for i in html_type:
        type.append(i.get_text())
        
    for i in html_name:
        name.append(i.get_text())
        
    for i in html_eng_name:
        eng_name.append(i.get_text())
        
    for i in html_degree:
        degree.append(i.get_text())

# step4.zip 모듈을 이용해서 list를 묶어주기        
list_sum = list(zip(type, name,eng_name,degree))


# step5.데이터프레임의 첫행에 들어갈 컬럼명
col = ['타입','이름','영어이름','도수']

# step6.pandas 데이터 프레임 형태로 가공
df = pd.DataFrame(list_sum, columns=col)

# step7.엑셀에 저장
df.to_excel('재료.xlsx')

print('end')
```
